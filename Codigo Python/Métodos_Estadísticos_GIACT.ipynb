```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Initialize necessary variables and data frames
# ...

# 1. Comparación de la Temperatura del Agua
temp_zona1 = np.random.normal(25, 2, 100)  # Temperatura en zona costera 1
temp_zona2 = np.random.normal(26, 2, 100)  # Temperatura en zona costera 2

# 2. Efecto de la Protección de Áreas Costeras en la Biodiversidad
biodiv_protected = np.random.normal(70, 10, 50)  # Biodiversidad en áreas protegidas
biodiv_unprotected = np.random.normal(60, 15, 50)  # Biodiversidad en áreas no protegidas

# 3. Influencia de la Salinidad en la Distribución de Especies
salinidad = np.random.uniform(30, 40, 100)  # Salinidad del agua
distribucion_especies = 0.5 * salinidad + np.random.normal(0, 2, 100)  # Distribución de especies

contaminacion_niveles = ['Bajo', 'Medio', 'Alto']
datos_arrecifes = pd.DataFrame({
    'Contaminacion': np.random.choice(contaminacion_niveles, 150, p=[0.5, 0.3, 0.2]),
    'Salud': np.random.choice(['Buena', 'Moderada', 'Mala'], 150, p=[0.5, 0.3, 0.2])
})

pH_primavera = np.random.normal(8, 0.5, 50)
pH_verano = np.random.normal(8.1, 0.5, 50)
pH_otonio = np.random.normal(7.9, 0.5, 50)
pH_invierno = np.random.normal(8, 0.5, 50)

profundidad = np.random.uniform(0, 100, 100)
abundancia_fitoplancton = 0.5 * profundidad + np.random.normal(0, 10, 100)

# 4. Prueba de Kruskal-Wallis para el impacto de la contaminación en la salud de los arrecifes de coral
health_mapping = {'Buena': 2, 'Moderada': 1, 'Mala': 0}
datos_arrecifes['Salud_numerica'] = datos_arrecifes['Salud'].map(health_mapping)

kruskal_stat, kruskal_p = stats.kruskal(*[
    datos_arrecifes[datos_arrecifes['Contaminacion'] == nivel]['Salud_numerica']
    for nivel in contaminacion_niveles
])

print(f"Prueba de Kruskal-Wallis para la salud de los arrecifes de coral: H = {kruskal_stat}, p = {kruskal_p}")

# 6. Regresión lineal para la relación entre la profundidad y la abundancia de fitoplancton
modelo_fitoplancton = sm.OLS(abundancia_fitoplancton, sm.add_constant(profundidad)).fit()

print(f"Regresión lineal para la abundancia de fitoplancton:\n{modelo_fitoplancton.summary()}")

# Plot the regression line
plt.scatter(profundidad, abundancia_fitoplancton)
plt.plot(profundidad, modelo_fitoplancton.params[0] + modelo_fitoplancton.params[1]*profundidad, color='red')
plt.xlabel('Profundidad')
plt.ylabel('Abundancia de fitoplancton')
plt.title('Regresión lineal entre la profundidad y la abundancia de fitoplancton')
plt.show()
```

```python
# Instalación de las bibliotecas necesarias en Google Colab
!pip install pymc3==3.11.4 arviz==0.11.4

import os

# Configuración para Theano en Google Colab
os.environ['MKL_THREADING_LAYER'] = 'GNU'
os.environ['THEANO_FLAGS'] = 'device=cpu,blas.ldflags=-lblas -lgfortran'

# Importación de las bibliotecas
import numpy as np
import pandas as pd
import pymc3 as pm
import arviz as az
import matplotlib.pyplot as plt

# Simulación de datos
np.random.seed(42)

# 1. Inferencia Bayesiana
supervivencia_corales = np.random.binomial(n=1, p=0.7, size=50)

with pm.Model() as modelo_inferencia:
    p = pm.Beta('p', alpha=1, beta=1)
    y = pm.Bernoulli('y', p=p, observed=supervivencia_corales)
    trace_inferencia = pm.sample(2000, return_inferencedata=True)

# 2. Modelos Jerárquicos Bayesianos
areas = np.repeat(np.arange(4), 25)
crecimiento_peces = np.random.normal(loc=5 + areas * 0.5, scale=1, size=100)

with pm.Model() as modelo_jerarquico:
    alpha = pm.Normal('alpha', mu=5, sigma=2)
    beta = pm.Normal('beta', mu=0.5, sigma=0.5)
    sigma = pm.HalfNormal('sigma', sigma=1)
    mu = alpha + beta * areas
    y = pm.Normal('y', mu=mu, sigma=sigma, observed=crecimiento_peces)
    trace_jerarquico = pm.sample(2000, return_inferencedata=True)

# 3. Regresión Bayesiana
temperatura_agua = np.random.uniform(20, 30, 50)
biomasa_algas = 2 * temperatura_agua + np.random.normal(0, 2, 50)

with pm.Model() as modelo_regresion:
    intercepto = pm.Normal('intercepto', mu=0, sigma=10)
    pendiente = pm.Normal('pendiente', mu=2, sigma=1)
    sigma = pm.HalfNormal('sigma', sigma=2)
    mu = intercepto + pendiente * temperatura_agua
    y = pm.Normal('y', mu=mu, sigma=sigma, observed=biomasa_algas)
    trace_regresion = pm.sample(2000, return_inferencedata=True)

# 4. Análisis de Supervivencia Bayesiano
tiempo_hasta_evento = np.random.exponential(scale=10, size=50)

with pm.Model() as modelo_supervivencia:
    lambda_ = pm.Gamma('lambda', alpha=2, beta=0.1)
    y = pm.Exponential('y', lam=lambda_, observed=tiempo_hasta_evento)
    trace_supervivencia = pm.sample(2000, return_inferencedata=True)

# 5. Análisis Espacial Bayesiano
coordenadas = np.random.uniform(0, 10, (30, 2))
biodiversidad_marina = np.sin(coordenadas[:, 0]) + np.cos(coordenadas[:, 1]) + np.random.normal(0, 0.5, 30)

with pm.Model() as modelo_espacial:
    l = pm.Gamma('l', alpha=2, beta=1)
    cov = pm.gp.cov.ExpQuad(2, ls=l)
    gp = pm.gp.Marginal(cov_func=cov)
    sigma = pm.HalfNormal('sigma', sigma=0.5)
    y_ = gp.marginal_likelihood('y', X=coordenadas, y=biodiversidad_marina, noise=sigma)
    trace_espacial = pm.sample(2000, return_inferencedata=True)

# Resultados
az.plot_trace(trace_inferencia, var_names=['p'])
az.plot_trace(trace_jerarquico, var_names=['alpha', 'beta'])
az.plot_trace(trace_regresion, var_names=['intercepto', 'pendiente'])
az.plot_trace(trace_supervivencia, var_names=['lambda'])
az.plot_trace(trace_espacial, var_names=['l', 'sigma'])
plt
```

```python
# Instalación de TensorFlow Probability en Google Colab
!pip install tensorflow-probability

import tensorflow_probability as tfp
import tensorflow as tf
import numpy as np

# Datos para inferencia bayesiana
supervivencia_corales = np.random.binomial(n=1, p=0.7, size=50).astype(np.float32)

# Modelo TensorFlow Probability para inferencia bayesiana
def model():
    p = tfp.distributions.Beta(concentration1=1., concentration0=1.)  # Prior para p
    return tfp.distributions.Bernoulli(probs=p)  # Likelihood

# Muestreo de Hamiltonian Monte Carlo
num_results = 2000
num_burnin_steps = 500
hmc = tfp.mcmc.HamiltonianMonteCarlo(
    target_log_prob_fn=lambda p: model().log_prob(supervivencia_corales).reduce_sum(),
    step_size=0.01,
    num_leapfrog_steps=3)
states, kernel_results = tfp.mcmc.sample_chain(
    num_results=num_results,
    num_burnin_steps=num_burnin_steps,
    current_state=[0.5],
    kernel=hmc)

# Resultados
p_trace = states[0].numpy()
print(f"Media posterior de p: {p_trace.mean()}")

```
