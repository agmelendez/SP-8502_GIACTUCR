```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import queue

# Central Limit Theorem Demonstration
def central_limit_theorem_demo(sample_size, num_samples):
    population = np.random.exponential(scale=5, size=10000)
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.choice(population, size=sample_size, replace=True)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    plt.hist(sample_means, bins=30, density=True, color='blue', alpha=0.7)
    plt.title('Central Limit Theorem Demonstration')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.show()

    print("Measures of Central Tendency for Sample Means:")
    print("Mean:", np.mean(sample_means))
    print("Median:", np.median(sample_means))
    print("Mode:", stats.mode(sample_means)[0][0])

# Queueing Theory Demonstration (M/M/1 Queue)
def queueing_theory_demo(lam, mu, num_customers):
    env = queue.Queue()
    arrival_times = np.random.exponential(scale=1/lam, size=num_customers)
    service_times = np.random.exponential(scale=1/mu, size=num_customers)

    for i in range(num_customers):
        env.enqueue((arrival_times[i], service_times[i]))

    wait_times = []
    while not env.is_empty():
        customer = env.dequeue()
        arrival_time, service_time = customer
        if env.is_empty():
            wait_time = 0
        else:
            wait_time = max(0, env.peek()[0] - arrival_time)
        wait_times.append(wait_time + service_time)

    print("\nQueueing Theory (M/M/1) Results:")
    print("Average Wait Time:", np.mean(wait_times))
    print("Average System Time:", np.mean(wait_times) + 1/mu)
    print("Utilization:", lam / mu)

# Main Function
if __name__ == "__main__":
    np.random.seed(42)

    # Central Limit Theorem Demonstration
    sample_size = 30
    num_samples = 1000
    central_limit_theorem_demo(sample_size, num_samples)

    # Queueing Theory Demonstration
    lam = 5  # Arrival rate
    mu = 7   # Service rate
    num_customers = 1000
    queueing_theory_demo(lam, mu, num_customers)
```
