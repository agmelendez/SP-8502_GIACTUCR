```python
import numpy as np
import pandas as pd

# Definimos el número de simulaciones
n_simulations = 1000

# Definimos probabilidades y distribuciones de las respuestas
probabilidades_conocimiento = np.random.beta(2, 5, 5)
probabilidades_percepcion = np.random.beta(3, 4, 5)
probabilidades_participacion = np.random.beta(4, 3, 2)

# Generamos datos simulados
simulacion_data = {
    'Nivel_Conocimiento': np.random.choice(np.arange(1, 6), n_simulations, p=probabilidades_conocimiento/np.sum(probabilidades_conocimiento)),
    'Percepcion_Valor': np.random.choice(np.arange(1, 6), n_simulations, p=probabilidades_percepcion/np.sum(probabilidades_percepcion)),
    'Participacion_Conservacion': np.random.choice([0, 1], n_simulations, p=probabilidades_participacion/np.sum(probabilidades_participacion)),
}

# Convertimos a DataFrame
df_simulacion = pd.DataFrame(simulacion_data)

# Mostramos el DataFrame de simulación
from IPython.display import display, Markdown
display(Markdown("## Simulación de Respuestas del Taller Participativo"))
display(df_simulacion)
```
